```python
class Hello:
    def hello(self):
        print('Hello World')
        
        
obj = Hello()
obj.hello()
```

```python
class Hello:
    @staticmethod
    def hello():
        print('Hello World')
        
        
obj = Hello()
obj.hello()
```

```python
class Cal:
    def cal(self,l,b):
        a=l*b
        print(a)

obj = Cal()
obj.cal(10,5)
```

```python
class Volume:
    def volume(self,l,b,h):
        v=l*b*h
        print(v)

class Area:
    def area(self,l,b):
        a=l*b
        print(a)

    
obj = Volume()
obj1 = Area()
obj.volume(10,5,2)
obj1.area(10,5)
```

```python
class Cal:
    def __init__(self,l,b,h):
        self.x=l
        self.y=b
        self.z=h
    
    def area(self):
        a=self.x*self.y
        print(a)
    def volume(self):
        v=self.x*self.y*self.z
        print(v)

obj = Cal(10,5,2)
obj.area()
obj.volume()
```

```python
class Info:
    def __init__(self,name,age,add):
        self.name=name
        self.age=age
        self.add=add
    
    def info(self):
        print(f"Hello i am {self.name}. I am from {self.add}. I am {self.age}")
    def data(self):
        print(self.name, self.age,self.add)
        
obj =Info("Ram",34,"Kathmandu")
obj.info()
obj.data()
```
